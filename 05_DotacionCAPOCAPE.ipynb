```python
lista_excel = ["Dotación_FRET_202507.xlsx","Dotación_FRET_202508.xlsx", "Dotación_FRET_202509.xlsx", "Dotación_FRET_202510.xlsx", "Dotación_FRET_202511.xlsx",
               "Dotación_FRET_202512.xlsx", "Dotación_FRET_202601.xlsx", "Dotación_FRET_202602.xlsx", "Dotación_FRET_202603.xlsx", "Dotación_FRET_202604.xlsx", "Dotación_FRET_202605.xlsx", "Dotación_FRET_202606.xlsx"]
ruta = r"input\bd_dotacion"
```

```python
df_dotacion_2024 = pd.read_excel(r"input\bd_dotacion\bk_Dotacion202401_202506.xlsx")
```

```python
rutas_completas = [os.path.join(ruta, archivo) for archivo in lista_excel]

# 2. Leer los DataFrames manteniendo la compresión de listas
dfs = [
    pd.read_excel(p, sheet_name="Detail_Capo", header=1) for p in rutas_completas
]

# 3. Concatenar usando 'keys' para inyectar el nombre del archivo (operación nativa en C)
df_dotacion = pd.concat(dfs, keys=lista_excel, ignore_index=False)

# 4. Pasar el nombre del archivo del índice a una columna y limpiar el MultiIndex
df_dotacion = df_dotacion.reset_index(level=0).rename(
    columns={"level_0": "Nombre_Archivo"}
)
df_dotacion = df_dotacion.reset_index(drop=True)
```

```python
df_dotacion[["Nombre_Archivo", 'Tipo dotación', 'Formato', 'Ubicación', 'Departamento',
       'Nombre de Posición', 'Categoría', 'Ceco',
       'Fecha', "Dotación Teórica Full", "Dotación Teórica Part", "Dotación Real Full", "Dotación Real Part"]]

id_columns = ["Nombre_Archivo",
        "Tipo dotación",
        "Formato",
        "Ubicación",
        "Departamento",
        "Nombre de Posición",
        "Categoría",
        "Ceco",
    ]

    # Columnas métricas que se colapsarán en una sola dimensión
value_columns = [
       "Dotación Teórica Full",
       "Dotación Teórica Part",
       "Dotación Real Full",
       "Dotación Real Part",
       "Vulnerables Full",
       "Vulnerables Part"
]

# Operación vectorizada de reshaping
df_long = pd.melt(
       df_dotacion,
       id_vars=id_columns,
       value_vars=value_columns,
       var_name="TipoDotacionx",
       value_name="HeadCount",
)
```

```python
mapeo_dotacion_invertido = {
    "Dotación Real Full": "RealFT",
    "Dotación Real Part": "RealPT",
    "Dotación Teórica Full": "TeoFT",
    "Dotación Teórica Part": "TeoPT",
    "Vulnerables Full": "VulneraFT",
    "Vulnerables Part": "VulneraPT",
}
```

```python
df_long["TipoDotacion"] = df_long["TipoDotacionx"].map(mapeo_dotacion_invertido)
```

```python
df_long[df_long["TipoDotacion"].isin(["Dotación Teórica Full", "Dotación Teórica Part"])]["HeadCount"].sum()
```

```python
df_long["_JEq"] = np.where(
    df_long["TipoDotacion"].str.contains("FT", na=False),
    df_long["HeadCount"],
    df_long["HeadCount"] / 2,
)
```
